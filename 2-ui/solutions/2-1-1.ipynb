```python
import sys
print(sys.executable)
from IPython.display import display
from ipywidgets import interact_manual

@interact_manual(
    name='',
    major=["IST", "IMT", "ADA"],
    gpa=(0.0, 4.0, 0.05)
)
def onclick(name, major, gpa):
    if gpa < 1.8:
        status = "Probation"
    elif gpa < 3.5:
        status = "Honors"
    else:
        status = "No List"
    
    display(f"{name} from {major} has a GPA of {gpa} and is on the {status}.")
```
